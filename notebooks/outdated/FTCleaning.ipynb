```python
import json
import pandas as pd
# Opening JSON file
f = open('../data/raw/CMS_2010_to_June_2022_ENGLISH.json')
# returns JSON object as
# a dictionary
data = json.load(f)
df = pd.DataFrame.from_dict(data)
```

```python
df['lastModifiedDate'] = pd.to_datetime(df['lastModifiedDate'])
```

```python
df.set_index('lastModifiedDate', inplace = True)
#df = df.loc['2015-06-07':'2016-01-01']
```

```python
import matplotlib.pyplot as plt
df.resample('1W').count()['id'].plot()
plt.title('Number of unique id per week')
```
